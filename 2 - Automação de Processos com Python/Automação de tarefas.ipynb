# Problema
1 buscar dados de valores
2 Criar as análises solicitadas pelo gestor
3 Enviar um email com o resultado das análises

# Parte 1 - Buscar dados

```python
!pip install yfinance
```

```python
import yfinance
```

```python
ticker = input('Digite o código da ação: ')
dados = yfinance.Ticker(ticker).history('6mo')
fechamento = dados.Close
fechamento.plot()
```

# Passo 2 - Criar análises
    
   Cotação máx
-Cotação min
-Cotação atual

```python
maxima = fechamento.max()
```

```python
minima = fechamento.min()
```

```python
atual = fechamento[-1]
```

# Passo 3 - Envio de email

- Abrir uma nova aba e digitar endereço de email
- Clicar no botão escrever novo email
- Preencher nome do destinatário  
- Preencher o assunto
- Escrever o email com os resultados da análise
- Clicar em enviar

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2

pyautogui.hotkey('ctrl', 't')
pyperclip.copy('www.hotmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('enter')
pyautogui.click(x=168, y=204)

pyperclip.copy('fabiolima.8@hotmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')

pyperclip.copy('Análises Diárias')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')
mensagem = f'''
Prezado Gestor,

Seguem as análises dos últimos seis meses da ação {ticker}:

Cotação máxima: {round(maxima, 2)}
Cotação mínima: {round(minima, 2)}
Cotação atual: {round(atual, 2)}

Qualquer dúvida, estou a disposição!

Atte.
'''
pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl', 'v')

pyautogui.click(x=331, y=268)






```
